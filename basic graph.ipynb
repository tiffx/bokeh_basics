```python
# Making a basic Bokeh line graph

# Importing Bokeh
from bokeh.plotting import figure
from bokeh.io import output_file, show

# Prepare some data
x = [1,2,3,4,5]
y = [6,7,8,9,10]
x2 = [3,7.5,10]
y2 = [3,6,9]

# Prepare the output file
output_file("Line.html")

# Create a figure object
f = figure()

# Create line plot
f.line(x,y)
show(f)

# Write the plot in the figure object
output_file("Triangle.html")
t = figure()
t.triangle(x2,y2)
show(t)

output_file("Circle.html")
c = figure()
c.circle(x2,y2)
show(c)

```

```python
import pandas
from bokeh.plotting import figure, output_file, show
 
p=figure(plot_width=500,plot_height=400, tools='pan')
 
p.title.text="Cool Data"
p.title.text_color="Gray"
p.title.text_font="times"
p.title.text_font_style="bold"
p.xaxis.minor_tick_line_color=None
p.yaxis.minor_tick_line_color=None
p.xaxis.axis_label="Date"
p.yaxis.axis_label="Intensity"    
 
p.line([1,2,3],[4,5,6])
output_file("graph.html")
show(p)
```

```python
import pandas
from bokeh.plotting import figure, output_file, show

df = pandas.read_csv("adbe.csv",parse_dates=["Date"])

p = figure(width=500,height=250,x_axis_type="datetime",sizing_mode='scale_width')

p.line(df["Date"],df["Close"],color="Orange",alpha=1)
p.circle(df["Date"],df["Open"],size=3,color="Blue",alpha=1)

output_file("Timeseries.html")
show(p)
```

```python
import pandas
from bokeh.plotting import figure, output_file, show
from bokeh.models import HoverTool

df = pandas.read_csv("Sample_of_the_produced_time_values.csv")

p = figure(width=500,height=100,x_axis_type="datetime",sizing_mode='scale_width',title="Motion Graph")
p.yaxis.minor_tick_line_color = None
p.ygrid[0].ticker.desired_num_ticks=1

hover = HoverTool(tooltips=[("Start ", "@Start{%d %m %H:%M:%S}"),("End ", "@End{%d %m %H:%M:%S}")],formatters={"Start": "datetime","End": "datetime"})
p.add_tools(hover)

p.quad(left=pandas.to_datetime(df["Start"]),right=pandas.to_datetime(df["End"]),bottom=0,top=1,color="Green")

output_file("MotionGraph.html")
show(p)
```
